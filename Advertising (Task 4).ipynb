```python
df.isna().sum()/100
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(df[['TV', 'Radio', 'Newspaper', 'Sales']])
plt.xlabel('Outliers Of [TV, Radio, Newspaper, Sales]')
plt.ylabel('Values Ranging')
plt.title('Distribution of Advertising Components')
plt.grid(alpha=0.3)
plt.show()
```

## 3. Data Visualisation

```python
# Let's see how Sales are related with other variables using scatter plot.
sns.pairplot(df, x_vars=['TV', 'Newspaper', 'Radio', 'Sales'], y_vars='Sales', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
# Let's see the correlation between different variables.
sns.heatmap(df.corr(), cmap="YlGnBu", annot = True)
plt.show()
```

```python
df_top_20 = df.head(20)

# Plotting the bar chart for the top 20 indices
plt.figure(figsize=(10, 6))
plt.bar(df_top_20.index, df_top_20['Newspaper'], width=0.4, label='Newspaper', align='center')
plt.bar(df_top_20.index + 0.4, df_top_20['Sales'], width=0.4, label='Sales', align='center')

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Top 20 Advertising and Sales: Newspaper vs. Sales')
plt.xticks(df_top_20.index)
plt.legend()
plt.show()
```
