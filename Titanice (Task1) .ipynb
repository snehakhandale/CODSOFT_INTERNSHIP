### 3. Handling Missing Data

```python
df.isna().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
fillna = df['Age'].mean()
print(fillna)
```

```python
df
```

```python
df.dropna(subset=['Embarked'], inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df.fillna({'Cabin': 'Unknown'}, inplace=True)
df['Cabin'] = df['Cabin'].fillna('Unknown')
```

```python
df['Cabin'].unique()
```

```python
df.isna().sum()
df
```

### 4. Filter values using 2 columns

```python
# Filter rows where 'text_column' equals a specific value
df[(df['Sex'] == 'male') & (df['Fare'] >= 7.2500)]
# print(filtered_df)
```

### 5. Using groupby calculate mean of age column based on Sex

```python
# Group by 'category_column' and calculate mean of 'numeric_column'
grouped = df.groupby('Sex')['Age'].mean()
print(grouped)
```

### 6. Filter dataset based on values and groupby them based on Name and find mean for Fare

```python
#filter dataset based on cabin and groupby based on name and mean for fare column.

filter_df = df[df['Cabin'] == 'Unknown']
filter_df
```

```python
groupby_filterdata = filter_df.groupby('Name')['Fare'].mean()
print(groupby_filterdata)
```

### 7. Aggregation using more than 2 columns with groupby

```python
#Now groupby and perform aggregations of your own columns which should be more than two.

agg_data = df.groupby('Sex').agg({ 'Age' : 'mean' , 'SibSp' : 'mean', 'Fare' : 'mean', 'Pclass' : 'mean'})
print(agg_data)
```

### 8. Distribution of Age

```python
plt.figsize=((10,6))
sns.histplot(data=df, kde=True, bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
plt.show()
```

### 9. Survival Rate by Gender

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

### 10. Passenger Class Distribution

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Count')
plt.ylabel('Count')
plt.show()
```

### 11. Relationship Between Age and Fare

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Relationship Between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

### 12. Survival Rate by Passenger Class and Gender

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=df)
plt.title('Survival Rate by Passenger Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

### 13. Embarked Location and Survival

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival Rate by Embarkation Location')
plt.xlabel('Embarked Location')
plt.ylabel('Count')
plt.show()
```

### 14. SibSp vs. Parch

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SibSp', y='Parch', data=df)
plt.title('SibSp vs. Parch')
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Number of Parents/Children Aboard')
plt.show()
```

### 15. Gender Distribution


```python
gender_counts = df['Sex'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.show()
```
